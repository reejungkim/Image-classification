```python
link_img = 'https://www.kareit.or.kr/uploaded/summernote/202205/c9ea9ef8c99914ced9e149d6255d087a.png'
#'https://kareit.or.kr/uploaded/summernote/202205/716f4b3ea695f9f44f56fdc22f74f882.png'
```

```python
from PIL import Image
from pytesseract import pytesseract

import requests
from io import BytesIO
  
response = requests.get(link_img)
img = Image.open(BytesIO(response.content))
 
```

```python
#tesseract의 추출 퀄리티를 높이기 위해 이미지 사이즈 확대
img = img.resize((img.size[0]*3, img.size[1]*3) ) 
img
```

 cd / && sudo find . -iname tesseract

```python
# Defining paths to tesseract.exe
path_to_tesseract = r'/usr/local/bin/tesseract'

# Providing the tesseract executable
# location to pytesseract library
pytesseract.tesseract_cmd = path_to_tesseract

# Passing the image object to image_to_string() function
# This function will extract the text from the image
text = pytesseract.image_to_string(img, 
                                   lang= 'KOR+ENG', #'KOR', 
                                   config='--psm 4 -c preserve_interword_spaces=1'+'\n')

# Displaying the extracted text
print(text[:-1])
```
